## **TRUE NEWS**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('True.csv')
data
```

```python
data.dropna(inplace = True)
data.isnull().sum()
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
```

```python
def display_cloud(data):
  cloud = WordCloud(
      background_color = 'white',
      stopwords = stopwords,
      max_words = 100,
      max_font_size = 30,
      scale = 5,
      random_state = 1)
  
  cloud = cloud.generate(str(data))
  fig = plt.figure(1,figsize=(12,12))
  plt.axis('off')
  plt.imshow(cloud)
  plt.show()

```

```python
display_cloud(data)
```

# **Full DataSet Exploration**

```python
data = pd.read_csv('Final_Clean.csv',encoding= 'unicode_escape')
data
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
from collections import Counter
import pandas as pd
import nltk

top_N = 11
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
data["unpunctuated"] = data['statement'].str.replace('[^\w\s]','')
```

```python
RE_stopwords = r'\b(?:{})\b'.format('|'.join(stopwords))
words = (data.unpunctuated.str.lower().replace([r'\|', RE_stopwords], [' ', ''], regex=True).str.cat(sep=' ').split())
rslt = pd.DataFrame(Counter(words).most_common(top_N),columns=['Word', 'Frequency']).set_index('Word')
rslt.plot.bar(rot=0, figsize=(16,10), width=0.8)                    
```
